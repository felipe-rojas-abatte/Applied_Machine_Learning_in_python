# Que es Machine Learning (ML)?

Es el estudio de programas de coputacion (algritmos) que tengan la capacidad de aprender mediante la repetición.

Para resolver un problema dado podemos implementar un codigo en algun lenguaje de programación que sea capaz de, con algunos inputs que le demos, entregarnos una respuesta. Esto se ha usado desde hace mucho tiempo en diferentes disciplinas como ciencias, economía, etc. 
Ahora, ¿que ocurre cuando la complejidad del problema hace que sea dificil crear un algoritmo capaz de resolverlo? 
Por ejempo: crear un programa que sea capaz de reconocer la voz y digitalizar lo dicho en un texto. Este problema es bastante complejo dada la cantidad de diferentes idiomas y las diferentes formas que tiene la gente de hablar. Un código basado en algun algoritmo "clasico" sería muy complejo de escribir, con reglas muy específicas para cada tipo de persona y cada tipo de voz, lo cual haría al final del dís que no sería tan robusto. ML por otra parte, tiene la capacidad de automaticamente aprender estas reglas de forma eficiente a traves de un entrenamiento con ejemplos etiquetados. A esto se le llama training data. Estas reglas deberían ser capaces, una vez terminado el entrenamiento, de generalizar el problema, reconocer y predecir correctamente nuevos ejemplos que no estan en el set de entrenamiento.
El problema básico de ML es como hacer que el computador se programe a si mismo, para buscar este set de complejas reglas y así mejorar la eficiencia del programa a medida que va ganando mas experiencia.
Para esto ML necesita de herramientas de estadistica y computer science, que le permitan encontrar la respuesta al problema y discernir si la respuesta es la confiable o no.

# Conceptos clave en Machine Learning

Podemos categorizar los problemas de ML en 2 tipos:

a) De aprendizaje supervisado (Supervised learning): Aprende a precedir valores basados en los datos del entrenamiento. Dentro de esta categoria podemos resolver 2 tipos de problemas:

i) Clasificación: Los valores de salida son clases discretas. ej: predicción de fraudes bancarios en una transacción o una palabra de salida en el reconocimiento de voz, etc. Y la función que usamos se llama clasificador.

ii)Regresión: Los valores de salida son valores continuos. ej: la cantidad de segundos que le tomaría a un auto acelerar de 0 a 100 km/h. Usamos la función de regresión.

En el aprendizaje supervisado podemos visualizar como funciona de la siguiente forma. Tenemos un conjunto de datos que usaremos para entrenar el programa en donde tenemos 2 variables (inputs) X e (outputs) Y. Cada input tiene un único output. Lo que queremos hacer es encontrar la función (clasificadora) que mapee estas 2 variables, para luego poder hacer una predicción que no está en el conjunto de datos de entrenamiento. Ahora, si queremos hacer la predicción, tenemos que tener un conjunto de datos para entrenen el programa. ¿De donde viene este conjunto de datos?, generalmente son proporcionados por humanos. Debemos introducirlos a mano.

b) De aprendizaje no supervisado (Unsupervised learning): Encuentra estructura en datos no etiquetados, es decir, no tenemos datos para entrenar el programa. La ides es que el programa sea capaz de encontrar estructura dentro de los datos. Dentro de esta categoría podemos resolver 2 tipos de problemas:

i) Clustering: Encontrar grupos de instancias similares en los datos. ej: En una pagina de venta por internet, encontrar los grupos de consumidores para ciertos tipos de productos.

ii) Outlier detection: Encontrar algun patron inusual. ej. Ataque de hackers en una sitio web. No sabemos como ni por donde va a atacar el hacker, así que que el programa debería ser capaz de predecir ese ataque.
    
# Como aplicar Machine Learning para resolver problemas (Workflow)

Hay 3 pasos básicos que se deben seguir.

a) Representación: Debemos ser capaces de presentar el problema de una    forma en que el computador pueda entender. Saber que input debemos    introducir al programa y entender el tipo de clasificador usar para    el problema. ej: pixels de una imagen.
    
b) Evaluación: Debemos tener un metodo que nos diga si el                calsificador que usamos fue bueno o malo. Esto nos permitiría ver      a eficiencia de distintos calsificadores.

c) Optimizacion: Necesitamos buscar los mejores parametros o              configuraciones para que el calsificador sea más óptimo y nos de el    mejor resultado al problema.

# Herramientas para Machine Learning

Las librerias que usaremos para trabajar en ML son: scikit-learn, Scipy, NumPy, pandas y matplotlib.
La librería mas importante para ML es scikit-learn. Es una de las librerías open source más usadas y con mayor documentación online. 
Existen muchos tutoriales, ejemplos y documentación que podemos usar.

a) http://scikitlearn.org/stable/user_guide.html (guia)

b) http://scikitlearn.org/stable/modules/class.html 

Ambas contienen documentacion más profunda y distintos algotirmos con sus respectivas opciones de uso.

Scikit-learn usa otras librerías cientificas de python: SciPy y NumPy.

a) Scipy: es una librería que soporta manipulación de datos y métodos de análisis bastante usados en la comunidad científica, como algebra lineal, funciones matematicas, distribuciones estadisticas, etc.

b) NumPy: es una librería que soporta la estructira fundamental usada por scikit-learn, como arreglos multidimensionales. Generalmente los inputs de scikit-learn son en formato de arreglos.

c) Pandas: es una librería para la manipulación de datos.

d) matplotlib: es una librería para realizar gráficos en 2D.
