### Simple regression dataset
Aquí presentamos el codigo para realizar un simple scatter plot con 100 puntos. El eje x representa los valores futuros y el eje y representa el target de regresión. Mostramos 2 figuras donde podemos visualizar un problema de regresión simple, donde claramente se ve una tendencia de los datos y uno mas complejo, donde es dificil visualizar una tendencia clara. Para crear estos plots se usaron las funciones make_regresion y make_friendman1 en SK learned dataset.

```python
# synthetic dataset for simple regression
from sklearn.datasets import make_regression
plt.figure()
plt.title('Sample regression problem with one input variable')
X_R1, y_R1 = make_regression(n_samples = 100, n_features=1,
                            n_informative=1, bias = 150.0,
                            noise = 30, random_state=0)
plt.scatter(X_R1, y_R1, marker= 'o', s=50)
plt.show()

# synthetic dataset for more complex regression
from sklearn.datasets import make_friedman1
plt.figure()
plt.title('Complex regression problem with one input variable')
X_F1, y_F1 = make_friedman1(n_samples = 100,
                           n_features = 7, random_state=0)

plt.scatter(X_F1[:, 2], y_F1, marker= 'o', s=50)
plt.show()
```

### Clasificación Binatia
Acá mostramos un data set con 2 clases que podemos representar tambien en un scatter plot pero esta vez usamos un color para cada clase. Usamos la función make_classificatio en SK learned dataset. En este caso podemos ver que estas 2 clases estan separadas linealmente, es decir que usando un clasificador lineal entre ellos nos puede dar una buena eficiencia a la hora de predecir sus clases.

```python
# synthetic dataset for classification (binary) 
plt.figure()
plt.title('Sample binary classification problem with two informative features')
X_C2, y_C2 = make_classification(n_samples = 100, n_features=2,
                                n_redundant=0, n_informative=2,
                                n_clusters_per_class=1, flip_y = 0.1,
                                class_sep = 0.5, random_state=0)
plt.scatter(X_C2[:, 0], X_C2[:, 1], c=y_C2,
           marker= 'o', s=50, cmap=cmap_bold)
plt.show()
```

### Clasificación binaria más compleja
En este caso tambien tenemos una conjunto de datos distribuidos en 2 clases, pero esta vez podemos ver que una clasificación lineal no apropiada, más bien podemos ver que se forman clusters entre los datos. Estos datos fueron usados en 2 pasos: 
- Primero usando la función make_blobs para generar 100 muestras en 8 diferentes clusters
- Y luego cambiando la etiqueta de cada cluster pasando de una etiqueta que va desde 1 a 8 a una binaria con solo 2 clases.

```python
# more difficult synthetic dataset for classification (binary) 
# with classes that are not linearly separable
X_D2, y_D2 = make_blobs(n_samples = 100, n_features = 2, centers = 8,
                       cluster_std = 1.3, random_state = 4)
y_D2 = y_D2 % 2
plt.figure()
plt.title('Sample binary classification problem with non-linearly separable classes')
plt.scatter(X_D2[:,0], X_D2[:,1], c=y_D2,
           marker= 'o', s=50, cmap=cmap_bold)
plt.show()
```

### Regression
K-Nearest neighbor también puede usarse para modelos de tipo regresión. Dado que no tenemos clases en esti tipo de dataset, sino que datos continuos, lo que hacemos es considerar el o los vecimos (dependiendo del valor de k que escogemos) del eje x más cercano del punto que queremos evaluar y calculamos el promedio de esos puntos para encontrar el valor target que vamos a predecir.

En este caso usamos la clase KNeighborsRegressor para entrenar, donde debemos indicar el numero n de vecinos a considerar en el modelo.

Dado que los valores a usar en el caso de regresión son continuos en comparación con los valores discretos de modelo de clasificación, si queremos ver que tam bien el modelo trabaja debemos usar otro indicador llamad R^2 (r-cuadrado) o coeficiente de determinación. Este indicador entrega valores enter 0.0 y 1.0 donde:
- 0 corresponde a un modelo constante donde todos los valores de predicción corresponden al promedio de los valores de entrenamiento.
- 1 corresponde a una predicción perfecta.

```python
from sklearn.neighbors import KNeighborsRegressor

X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1, random_state = 0)

knnreg = KNeighborsRegressor(n_neighbors = 5).fit(X_train, y_train)

print(knnreg.predict(X_test))
print('R-squared test score: {:.3f}'
     .format(knnreg.score(X_test, y_test)))
```

```python
fig, subaxes = plt.subplots(1, 2, figsize=(8,4))
X_predict_input = np.linspace(-3, 3, 50).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X_R1[0::5], y_R1[0::5], random_state = 0)

for thisaxis, K in zip(subaxes, [1, 3]):
    knnreg = KNeighborsRegressor(n_neighbors = K).fit(X_train, y_train)
    y_predict_output = knnreg.predict(X_predict_input)
    thisaxis.set_xlim([-2.5, 0.75])
    thisaxis.plot(X_predict_input, y_predict_output, '^', markersize = 10,
                 label='Predicted', alpha=0.8)
    thisaxis.plot(X_train, y_train, 'o', label='True Value', alpha=0.8)
    thisaxis.set_xlabel('Input feature')
    thisaxis.set_ylabel('Target value')
    thisaxis.set_title('KNN regression (K={})'.format(K))
    thisaxis.legend()
plt.tight_layout()
```

Veamos como cambia la complejidad de la predicción como función de k.
Tomemos algunos valores de k como 1, 3, 5, 7, 15, 55.
Valores más bajos de k entrega modelos con mayor complejidad como podemos ver en la figura con k=1. El valor de R^2 en este caso es 1.0 para el training sample, pero es muy malo para predecir el test sample con un R^2 = 0.155. 
A medida que aumentamos k podemos ver que la regresiṕn se hace más suave aumentando el valor de R^2 para la predicción del test sample. El modelo con k=15 presenta la mejor predicción, pero para k=55 esta vuelve a disminuir, haciendo que el modelo sea muy simplista para la predicción (underfit).

```python
# plot k-NN regression on sample dataset for different values of K
fig, subaxes = plt.subplots(5, 1, figsize=(5,20))
X_predict_input = np.linspace(-3, 3, 500).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X_R1, y_R1,
                                                   random_state = 0)

for thisaxis, K in zip(subaxes, [1, 3, 7, 15, 55]):
    knnreg = KNeighborsRegressor(n_neighbors = K).fit(X_train, y_train)
    y_predict_output = knnreg.predict(X_predict_input)
    train_score = knnreg.score(X_train, y_train)
    test_score = knnreg.score(X_test, y_test)
    thisaxis.plot(X_predict_input, y_predict_output)
    thisaxis.plot(X_train, y_train, 'o', alpha=0.9, label='Train')
    thisaxis.plot(X_test, y_test, '^', alpha=0.9, label='Test')
    thisaxis.set_xlabel('Input feature')
    thisaxis.set_ylabel('Target value')
    thisaxis.set_title('KNN Regression (K={})\n\
Train $R^2 = {:.3f}$,  Test $R^2 = {:.3f}$'
                      .format(K, train_score, test_score))
    thisaxis.legend()
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```

### Parametros importantes en los modelos del tipo KNeighbors Classifier and KNeighbors Regressor
- Complejidad del modelo: n° de vecinos k a cponsiderar. Por defecto se consideran 5
- Fiteo del modelo: Metrica. DIstancia entre puntos en un espacio plano (metrica de Minkowski)
